```python
# Funtion to display image

def display (img):
    tmp = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.figure(figsize=(12,10))
    plt.imshow(tmp)
    plt.show()
```

```python
query_image = cv.imread('query.png')
display(query_image)
```

```python
# Open dataset
database_images = glob.glob('dataset/*')
image_test = len(database_images)
database_images = database_images[:image_test]

# Test result
test_results = []

# Run test on every image
for filename in database_images:
    img = cv.imread(filename)
    
    sample_match_1_result = sampleMatch1(query_image, img)
    sample_match_2_result = sampleMatch2(query_image, img)
    
    similarity_value = sample_match_1_result * 0.5 + sample_match_2_result * 0.5;
    test_results.append((similarity_value, filename, img))
    #display(image)
```

```python
# Take 5 most similar image
test_results.sort(key = lambda tup:tup[0], reverse=True)
test_results_sorted = test_results[:5]

for (value, filename, image) in test_results_sorted:
    display(image)
    print('filename : ', filename)
    print('similarity value : ', value)
```
